Travail sur les catégorie des délais de livraisons


```python
#modification de mon Champ en TIMEDELTA

df_W_1['dif_livre_estime_2'] = pd.to_timedelta(df_W_1['dif_livre_estime_2'], errors='coerce')
```

```python
print(df_W_1[['dif_livre_estime_2']].info())
```

```python
# Définir les catégories

# Fonction de catégorisation
def categoriser_delai(delta):
    if delta < pd.Timedelta(days=1):
        return "Livraison Express (<1 jour)"
    elif delta < pd.Timedelta(days=3):
        return "Rapide (1-3 jours)"
    elif delta < pd.Timedelta(days=7):
        return "Standard (3-7 jours)"
    else:
        return "Longue (>7 jours)"

# Appliquer la fonction sur la colonne
df_W_1['categorie_livraison'] = df_W_1['dif_livre_estime'].apply(categoriser_delai)

# Affichage du résultat
print(df_W_1)


```

```python
df_W_1.head(3)
```

```python
df_W_1['categorie_livraison'].value_counts()
```

```python
# Répartition des livraison en catégorie de livraisons


# Calculer les valeurs
counts = df_W_1['categorie_livraison'].value_counts()
labels = counts.index
sizes = counts.values

# Créer le camembert avec légendes et pourcentages
plt.figure(figsize=(8, 6))
plt.pie(
    x=sizes,
    labels=[f'{label} ({size})' for label, size in zip(labels, sizes)],  # Ajoute le compte dans les labels
    autopct='%1.1f%%',  # Affiche le pourcentage
    startangle=90,  # Angle de départ pour une meilleure lisibilité
    colors=plt.cm.Paired.colors  # Palette de couleurs harmonieuse
)
plt.title('Répartition des livraisons par catégorie de délai')
plt.show()

```
