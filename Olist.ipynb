***Quel produit est le plus vendu?***

```python
# Fusionner les donn√©es des commandes et des articles de commande
merged_data = order_items.merge(products, on='product_id')
```

```python
merged_data
```

```python
# Compter le nombre de ventes par produit
product_sales = merged_data['product_id'].value_counts().reset_index()
product_sales.columns = ['product_id', 'sales_count']
```

```python
product_sales
```

```python
# Trouver le produit le plus vendu
most_sold_product = product_sales.iloc[0]
most_sold_product
```

```python
#trouver la cat√©gorie du produit le plus vendu :
product_id_most_sold = 'aca2eb7d00ea1a7b8ebd4e68314663af'
#filtrer la table produit pour obtenir la cat√©gorie
product_info = products[products['product_id'] == product_id_most_sold]
#r√©cup√©rer le nom de la cat√©gorie
product_category = product_info['product_category_name_english'].values[0]
product_category
```

***Quelles sont cat√©gorie de produit qui g√©n√®re le plus ventes?***

```python
#Compter le nombre de vente par cat√©gorie
category_sales = merged_data.groupby('product_category_name')['product_id'].count().reset_index()
category_sales.columns=['category', 'sales_count']
```

```python
category_sales
```

```python
most_sold_category = category_sales.sort_values(by='sales_count', ascending=False).iloc[0]
most_sold_category
```

```python
#Toute cat√©gorie
plt.figure(figsize=(10,14))
sns.barplot(x='sales_count', y='category', data=category_sales.sort_values(by='sales_count', ascending=False))
plt.title('Ventes par cat√©gorie de produit')
plt.xlabel('Nombre de ventes')
plt.ylabel('Cat√©gorie de produit')
plt.show()
```

```python
#top 5 des cat√©gorie max vente
plt.figure(figsize=(6,4))
sns.barplot(x='sales_count', y='category', data=category_sales.sort_values(by='sales_count', ascending=False).head())
plt.title('Ventes par cat√©gorie de produit')
plt.xlabel('Nombre de ventes')
plt.ylabel('Cat√©gorie de produit')
plt.show()
```

***Quelles est le chiffres d'affaire par produit ?***

```python
# Calculer le chiffre d'affaires par produit
merged_data['revenue'] = merged_data['price']  # Chaque ligne repr√©sente une vente, donc le prix est le chiffre d'affaires pour cette vente
revenue_per_product = merged_data.groupby('product_id')['revenue'].sum().reset_index()
```

```python
revenue_per_product
```

```python
# Trier par chiffre d'affaires - le produit qui √† g√©n√©r√© le plus de $
revenue_per_product = revenue_per_product.sort_values(by='revenue', ascending=False)
top_revenue_product = revenue_per_product.iloc[0]
top_revenue_product
```

```python

# Produit √† analyser
product_id_top_revenue = 'bb50f2e236e5eea0100680137654686c'

# Filtrer les ventes du produit dans 'order_items'
product_sales = order_items[order_items['product_id'] == product_id_top_revenue]

# Fusionner avec la table 'orders' pour obtenir la date d'achat
product_sales = product_sales.merge(orders[['order_id', 'order_purchase_timestamp']], on='order_id', how='left')

# Convertir la colonne en format datetime
product_sales['order_purchase_timestamp'] = pd.to_datetime(product_sales['order_purchase_timestamp'])

# Additioner les ventes par mois et calculer le chiffre d'affaires
product_sales['month'] = product_sales['order_purchase_timestamp'].dt.to_period('M')  # Regroupement par mois
monthly_revenue = product_sales.groupby('month')['price'].sum().reset_index()

# Convertir en datetime pour l'affichage sur le graphique
monthly_revenue['month'] = monthly_revenue['month'].astype(str)  # Convertir en char pour √©viter les erreurs

# Tracer le graphique
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_revenue, x='month', y='price', marker='o')

# Param√™tre du graphique
plt.title(f"√âvolution du chiffre d'affaires du produit {product_id_top_revenue} de 08-2017 √† 08-2028", fontsize=11)
plt.xlabel("P√©riode (mois)", fontsize=12)
plt.ylabel("Chiffre d'affaires (BRL)", fontsize=12)
plt.xticks(rotation=45)  # Incliner les dates pour une meilleure lisibilit√©
plt.grid(True)

# Afficher le graphique
plt.show()
```
