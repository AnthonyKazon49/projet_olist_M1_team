# Constitution de ma table vendeur

```python
#Jointure de la table Sellers avec la table orders_items

table_1= sellers.merge(order_items, how = 'right')
```

```python
#Jointure de la table_1 avec la table orders

table_2= table_1.merge(orders, how = 'right')
```

```python
#Jointure de la table_2 avec la table order_payements

table_3 = table_2.merge(order_payements, how = 'right')
```

```python
#Jointure de la table_3 avec la table products

table_31= table_3.merge(products, how = 'right')
```

```python
#Jointure de la table_31 avec la table name_translation

table_4= table_31.merge(name_translation, how = 'right')
```

```python
#Convertir le champs order_approved_at en datatime et rajouter la colonne year et month à ma table

table_4["order_approved_at"] = pd.to_datetime(table_4["order_approved_at"], errors="coerce")

# Créer une nouvelle colonne avec l'année

table_4["year"] = table_4["order_approved_at"].dt.year
# Créer une nouvelle colonne avec l'année

table_4["month"] = table_4["order_approved_at"].dt.month

#Dictionnaire des mois
mois_dict = {
    1.0: "Janvier", 2.0: "Février", 3.0: "Mars", 4.0: "Avril", 5.0: "Mai", 6.0: "Juin",
    7.0: "Juillet", 8.0: "Août", 9.0: "Septembre", 10.0: "Octobre", 11.0: "Novembre", 12.0: "Décembre"
}

# Remplacement des chiffres par les noms des mois
table_4["month"] = table_4["month"].replace(mois_dict)
```

```python
#Grouper le nombre de vente par année

result = table_4.groupby("year", as_index=False).agg({
 "order_id": "count",
   "payment_value": "sum"
})

print(result)
```

```python
#Grouper le nombre de vente par mois

result1 = table_4.groupby("month", as_index=False).agg({
 "order_id": "count",
   "payment_value": "sum"
})

print(result1)
```

```python
#Distribution du nombre de commande et du chiffre d'affaires par année

# Définir la taille de la figure
plt.figure(figsize=(14, 5))

# Histogramme des ventes par année
plt.subplot(1, 2, 1)
sns.barplot(x=result["year"], y=result["order_id"], palette="Blues")
plt.xticks(rotation=45)
plt.title("Distribution des ventes par année")
plt.xlabel("Année")
plt.ylabel("Nombre de ventes")
# Histogramme du chiffre d'affaires par année
plt.subplot(1, 2, 2)
sns.barplot(x=result["year"], y=result["payment_value"], palette="Greens")
plt.xticks(rotation=45)
plt.title("Distribution du chiffre d'affaires par année")
plt.xlabel("Année")
plt.ylabel("Montant total des paiements")

# Titre global
plt.suptitle("Analyse des ventes et du chiffre d'affaires par année")

# Ajuster l'affichage
plt.tight_layout()
plt.show()
```

```python
#Distribution du nombre de commande et du chiffre d'affaires par mois

# Définir la taille de la figure
plt.figure(figsize=(14, 5))

# Histogramme des ventes par année
plt.subplot(1, 2, 1)
sns.barplot(x=result1["month"], y=result["order_id"], palette="Blues")
plt.xticks(rotation=45)
plt.title("Distribution des ventes par mois")
plt.xlabel("mois")
plt.ylabel("Nombre de ventes")
# Histogramme du chiffre d'affaires par année
plt.subplot(1, 2, 2)
sns.barplot(x=result1["month"], y=result["payment_value"], palette="Greens")
plt.xticks(rotation=45)
plt.title("Distribution du chiffre d'affaires par année")
plt.xlabel("mois")
plt.ylabel("Montant total des paiements")

# Titre global
plt.suptitle("Analyse des ventes et du chiffre d'affaires par mois")

# Ajuster l'affichage
plt.tight_layout()
plt.show()
```
